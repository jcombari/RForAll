## 📊 R 🖼️ | Escanea múltiples imágenes a un solo PDF 🖼️➡️📄  

## ❓ ¿Tienes varias imágenes escaneadas que necesitas enviar como un solo archivo? 🙃  
Con R y `magick` puedes crear un PDF ordenado, local y seguro.  

## 👉 Solución  
📸 Genera múltiples imágenes en escala de grises (como si fueran escaneos)  
📄 Inserta cada imagen como página en un PDF tamaño A4 usando `grid` y `pdf()`  
🧾 Escala y centra cada imagen para una presentación profesional  

## 🔧 ¿Cómo funciona?  
🖼️ Creamos imágenes aleatorias en escala de grises para simular escaneos  
📄 Las guardamos localmente con un prefijo organizado  
🧾 Creamos un PDF con cada imagen escalada y centrada en páginas A4

```r
# 🔇 Silently install required packages
packages <- c("magick", "gridExtra", "grid")
to_install <- packages[!packages %in% rownames(installed.packages())]
if (length(to_install) > 0) {
  install.packages(to_install, quiet = TRUE)
}

# 🔇 Load packages suppressing messages, warnings, and system output
suppressWarnings(suppressMessages({
  try(library(magick), silent = TRUE)
  try(library(gridExtra), silent = TRUE)
  try(library(grid), silent = TRUE)
}))

# Function to create a circular icon with a black border
create_icon <- function(color, size = 50) {
  img <- image_blank(width = size, height = size, color = "none")
  img <- image_draw(img)
  symbols(x = size/2, y = size/2, circles = size/2 - 3, inches = FALSE,
          add = TRUE, bg = color, fg = "black", lwd = 3)
  dev.off()
  return(img)
}

# 📁 Configuration
prefix <- "20250620_"
directory <- paste0(prefix, "images")  # Folder "20250620_images"

# Create directory if it doesn't exist
if (!dir.exists(directory)) {
  dir.create(directory)
}

# 🖼️ Generate and save 3 images with text + icon
image_list <- c()

texts <- c(
  "Page 1: This is a sample scanned receipt.",
  "Page 2: Handwritten notes on data science.",
  "Page 3: Meeting minutes and agenda."
)

# Icon colors
icon_colors <- c("red", "green", "blue")

for (i in seq_along(texts)) {
  # Create a white base image
  img <- image_blank(width = 350, height = 250, color = "white")
  
  # Add text
  img <- image_annotate(img, texts[i], size = 15, color = "black", location = "+10+120")
  
  # Create a small circular icon
  icon <- create_icon(icon_colors[i], 50)
  
  # Composite: paste the icon in the top-right corner (+290,+10)
  img <- image_composite(img, icon, offset = "+290+10")
  
  # Save the image
  filename <- file.path(directory, paste0(prefix, "image_", i, ".png"))
  image_write(img, path = filename, format = "png")
  
  image_list <- c(image_list, filename)
}

# Check if images were created
if (length(image_list) == 0) stop(paste0("No PNG images with prefix '", prefix, "' found in '", directory, "'."))

# 📄 Create a PDF with the images (centered on A4)
pdf_filename <- file.path(directory, paste0(prefix, "combined_output.pdf"))

# Suppress printed page numbers while generating the PDF
invisible(pdf(pdf_filename, width = 8.27, height = 11.69))  # A4 size in inches

for (img_path in image_list) {
  img <- image_read(img_path)
  
  # Get image dimensions in pixels
  info <- image_info(img)
  img_w <- info$width
  img_h <- info$height
  
  # A4 dimensions in pixels at 72 dpi for PDF device
  a4_w <- 8.27 * 72
  a4_h <- 11.69 * 72
  
  # Scale the image to fit within A4 while maintaining aspect ratio
  scale_factor <- min(a4_w / img_w, a4_h / img_h)
  new_w <- img_w * scale_factor
  new_h <- img_h * scale_factor
  
  # Resize image
  img_scaled <- image_scale(img, paste0(round(new_w), "x", round(new_h)))
  raster_img <- as.raster(img_scaled)
  
  # Draw centered image
  grid.newpage()
  pushViewport(viewport(width = unit(new_w / a4_w, "npc"),
                        height = unit(new_h / a4_h, "npc"),
                        x = 0.5, y = 0.5, just = "center"))
  grid.raster(raster_img)
  popViewport()
}

dev.off()

# 📊 Display images side-by-side with spacing
plots <- lapply(image_list, function(p) {
  img <- image_read(p)
  rasterGrob(as.raster(img), interpolate = FALSE)
})

grid.arrange(grobs = plots, ncol = length(plots), padding = unit(1, "cm"))
```
